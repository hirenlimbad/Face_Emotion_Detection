```python
import cv2
from tensorflow import keras
import numpy as np

# Load the emotion detection model outside the loop
model = keras.models.load_model("CV_Face_Emotion")

def detect_faces(frame):
    face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
    
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5)

    return faces

def detect_emotion(face_frame):
    # Resize the face frame to 48x48 and convert to grayscale
    face_frame = cv2.resize(face_frame, (48, 48))
    face_frame = cv2.cvtColor(face_frame, cv2.COLOR_BGR2GRAY)
    face_frame = face_frame.reshape(-1, 48, 48, 1)
    face_frame = face_frame / 255.0

    emotions = ['angry', 'disgust', 'fear', 'happy', 'neutral', 'sad', 'surprise']
    
    # Use the loaded model to make predictions
    predictions = model.predict(face_frame)
    emotion_index = np.argmax(predictions)
    emotion = emotions[emotion_index]
    
    return emotion

cap = cv2.VideoCapture(0)  # 0 is the default camera

while True:
    ret, frame = cap.read()
    if not ret:
        break

    faces = detect_faces(frame)

    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)
        
        # Extract the small face frame
        face_frame = frame[y:y + h, x:x + w]
        
        # Pass the small face frame to the emotion detection method
        emotion = detect_emotion(face_frame)
        
        # Display the detected emotion for each face
        cv2.putText(frame, f"Emotion: {emotion}", (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)

    cv2.imshow('Face Detection', frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
